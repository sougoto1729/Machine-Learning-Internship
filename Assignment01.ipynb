```python
# 01

def factorial(num):
    if (num == 0 or num == 1):
        return 1
    return num * factorial(num - 1)


number = int(input("Enter the number: "))
print(f"Factorial of {number} is {factorial(number)}")
```

```python
# 03

num = int(input("Enter the number: "))
digits = 0
while (num != 0):
    digits += 1  # counting the last digit
    num //= 10  # removing the last digit
print("number of digits in the number is: ", digits)
```

```python
# 04

# !5 * !4 * !3 * !2 * !1 = 5^1 * 4^2 * 3^3 * 2^4 * 1^5

num = int(input("Enter the number: "))
p = 1
result = 1
while (num > 0):  # time complexity = O(n)
    result *= num ** p
    p += 1
    num -= 1
print("result:", result)
```

```python
# 05

def square_sum(*numbers):
    sum = 0
    for number in numbers:
        sum += number ** 2
    return sum


print(square_sum(1, 2, 3, 4))
```

```python
# 06

def do_operation(num1, num2, operator):
    if (operator == '+'):
        return num1 + num2
    elif (operator == '-'):
        return num1 - num2
    elif (operator == '*'):
        return num1 * num2
    elif (operator == '/'):
        return num1 / num2


print(do_operation(3, 2, '+'))
```

```python
# 07

def my_determination(times):
    verb1 = 'CAN'
    verb2 = 'WILL'
    for i in range(times):
        print(f"I {verb1}")
        verb1, verb2 = verb2, verb1  # swapping verb1 and verb2 in each iteration


my_determination(6)
```

```python
# 08

glove_list = [1, 2, 1, 2, 1, 3, 2]
n = len(glove_list)
freq = [-1] * n
for i in range(n):
    count = 1
    for j in range(i + 1, n):
        if (glove_list[i] == glove_list[j]):  # found a duplicate
            glove_list[j] = 0  # making the duplicates 0
            count += 1  # counting the duplicates
    if (glove_list[i] != 0):  # if the position was not containing the duplicate before
        freq[i] = count  # storing the frequency of the number in freq[i]
# in list freq we have successfully stored the frequency of the elements in glove_list
pairs = 0
for i in range(n):
    if (glove_list[i] != 0):
        # for glove_list[i], possible number of pairs = freq[i] // 2 where freq[i] stores the frequency of glove_list[i]
        pairs += freq[i] // 2

print("possible pairs:", pairs)
```

```python
# 09

def consecutive_combo(list1, list2):
    list = list1 + list2
    list.sort()
    isConsecutive = True
    for i in range(1, len(list)):
        if (list[i] - list[i - 1] != 1):
            isConsecutive = False
            break
    return isConsecutive


consecutive_combo([1, 4, 6, 5], [2, 3, 7, 8, 10])
```

```python
# 10

def profit(price_info):
    return round((price_info['sell_price'] - price_info['cost_price']) * price_info['inventory'])


profit(
    {
        'cost_price': 32.67,
        'sell_price': 45.00,
        'inventory': 1200
    }
)
```
